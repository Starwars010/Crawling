```python
#Bugs Music을 스스로 Crawling 해보자
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
#필요한 라이브러리들을 다 가져왔다
#이제 사이트 url을 가져와서 코드로써 의미를 부여해 보자
```

```python
url = "https://music.bugs.co.kr/chart"
html = requests.get(url) #url로 웸을 가져와라!
soup = bs(html.text, 'html.parser') #url의 코드 text화 시킨 것을 코드로써 의미있게 만들어라 #soup = bs(html.text)로만 써도 됨
```

```python
#이제는 select함수를 써서 태그를 가져와서 단어만 가져오기 즉 크롤링을 본격적으로 해보자
#소스코드를 보기 위해서는 사이트에 가서 오른쪽 클릭 하고 맨 아래 '검사' 옵션으로 가자
#soup는 전체 페이지의 코드니까 다듬어야할 필요가 있겠지
#일단은 100곡에 대한 정보(칸, 순위, 색깔, 글씨, 가수, 앨범, 노래제목 등등)을 가져와보자
#너무 바로 narrow down 해버리면 반복문 쓸 수 없으니까 100곡의 정보가 어딨는지만 보자구
songs_100 = soup.select('tr')
#tr 태그에 뭔가 100곡스러운 리스트가 있었던 것 같았다
#개수를 세보자
len(songs_100)
```

```python
#으잉? 104개네...전체 페이지에 100곡 말고도 표현된 tr 태그가 있나보자
#어딨을까?
#찾아봤더니 100곡 저 밑에 
#<section class="layer contents"> 있었고
#그 밑에 소속된 <table class="formlayout">
#밑에 <tbody>
#밑에 <tr> 한 3개 쯤 있었음
songs_100 = soup.select('tbody > tr')
len(songs_100)
```

```python
#전체 페이지에서 tbody에 속한 tr은 103개란 소리
#위에서 새로운 tr찾은 숫자랑 똑같으니까 make sense
#더 줄여보자
#100곡은 <table class="list trackList byChart"> 포함되어 있으니까 여기까지 올라가보자
#왜냐면 나머지 3곡은 <table class="formlayout">에 포함되어 있지 <table class="list trackList byChart">여기에 있지 않기 때문에 분별이 가능할 것

#이 이하는 다 됨...왜 꼭 부분으로만 해야 가능하져?? 더 부정확한 거 아닌가요??
#songs_100 = soup.select('table.trackList > tbody > tr')
#len(songs_100)
#songs_100 = soup.select('table.list > tbody > tr')
#len(songs_100)
songs_100 = soup.select('table.byChart > tbody > tr')
len(songs_100)

#이 이하는 왜 안되는겨?? 전체 이름으로 하면 왜 안되는겨?? 겹치는게 있으면 어쩌려구
#songs_100 = soup.select('table.list trackList > tbody > tr')
#len(songs_100)
#songs_100 = soup.select('table.listtrackList > tbody > tr')
#len(songs_100)
```

```python
#뭐...어쨋든 100곡으로 만들긴 했네요...
#그렇다면 이제 제목만 뽑아야 하니까 narrow it down해야할 때가 왔어요

#한 곡에 대한 방대한 정보들 중 딱 제목만 담고 있는 태그를 찾아봐야겠어요
#보니까 a태그에 <p class="title"> 안에 딱 제목만 있네요
songs_100_title = soup.select('p.title > a')
print(len(songs_100_title))
print(songs_100_title)
```

```python
#그런데 말이죠 p.title > a에 있는 제목은 100개가 된답니다 이 바보 멍청아...
#이 중에서 하나만 찾고 싶어요
#그러면 첫번째꺼가 뭔지 알고 싶었자나요??
songs_1_title=soup.select('p.title > a')[0]
print(songs_1_title)
```

```python
#우와 드디어 하나가 떳어요!
#근데 쓰잘데기 없는 코드같은 거는 버리고 예쁘게 글씨만 보였으면 좋겠어요.
#그러면 text의미가 있는 거만 뽑으면 대지용
#text는 사실상 blessed-cursed하나밖에 없네요
songs_1_title_text=soup.select('p.title > a')[0].text
print(songs_1_title_text)
```

```python
#어머! 드디어 됐어요!
#그러면 이제 가수도 알아봐야겠어요
#가수는 a태그에 <p class="artist">라고 표기돼어 있네요
songs_100_singer = soup.select('p.artist > a')
print(songs_100_singer)
```

```python
#그랬더니 가수 100명의 이름이 떳네요
#이제 1순위 가수를 보고 싶어요
songs_1_singer = soup.select('p.artist > a')[0]
print(songs_1_singer)
```

```python
#1위 곡 가수를 찾았어요
#이제 예쁘게 글씨만 보이게 할게요
songs_1_singer_text = soup.select('p.artist > a')[0].text
print(songs_1_singer_text)
```

```python
#우왕 이제 다 됐어요
#이제 for문을 활용해 100개의 곡과 가수를 추출해 볼게요
#근데 하나만 더 항목 추가해서 해 볼까요??음...앨범명까지 해봐요
#무슨 태그 밑에 앨범이름이 숨겨져있으까요??
#근데 얘는 좀 이상하네요
#원래 <td class="left">
#밑에 <p class='artist'>
#밑에 <a>
#태그가 있었는데 
#앨범명은 좀 달라요ㅠㅠ
#<td class="left">
#밑에 바로 <a>가 와버리고 같은 줄에 class='album'이라고 돼어있어요
#같은 거겠죠??함 해보죠
songs_100_album=soup.select('td.left > a.album')
print(songs_100_album)
```

```python
#미쳤네요~~당신이 해냈네요
#100곡의 앨범명이 보여요
#그럼 이제 1위 곡의 앨범명을 알아봐야쥬
songs_1_album=soup.select('td.left > a.album')[0]
print(songs_1_album)
```

```python
#우왕 하나만 나왔어요
#이제 쓰잘데기 없는 코드들 빼고 이쁘게 이름만 뺴봐요
songs_1_album_text=soup.select('td.left > a.album')[0].text
print(songs_1_album_text)
```

```python
#우왕 드디어 또 하셨네요~
#근데 이쯤 돼니까 애초에 soup이 모였더라 하쥬??
#soup이 뭐였나면요 전체 페이지 전체 코드 였지요~
#그래서 select함수를 통해서 narrow it down 한 거였자나요!!
#네네~잊을 수 있쥬
```

```python
#자~이제 그럼 for 문을 통해 이중 리스트에 항목들을 넣어봐야겠어요
bugs_music_list=[]
#먼저 빈 리스트를 생성해주죠
rank=1

for i in range(len(songs_100)):
    songs_1_title_text=soup.select('p.title > a')[i].text
    songs_1_singer_text=soup.select('p.artist > a')[i].text
    songs_1_album_text=soup.select('td.left > a.album')[i].text
    bugs_music_list.append(['bugs', rank, songs_1_title_text, songs_1_singer_text, songs_1_album_text])
    rank +=1
    
bugs_music_list
    
```

```python
#미친뇬아 니가 해냈습니다~!
#이제는 이쁘게 만들고 싶어요
#이중리스트 몬 생겼어요
#그때 pandas import했던 거 기억나시나요??그걸 쓸 겁니다
```

```python
pretty_list=pd.DataFrame(bugs_music_list, columns=['음원사이트', '순위', '제목', '아티스트', '앨범이름'])
pretty_list
```

```python
#이 이쁜 리스트를 코딩 사이트로만 남겨서 되겠어요??
#엑셀로 뽑아보쥬
#이때까지는 인덱스 표시를 안했으니 이번에는 인덱스 표시까지 해보죠
pretty_list.to_excel('bugs_music_top100.xlsx', index=True)
```
