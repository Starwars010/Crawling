```python
import selenium
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
#드라이버 불러오기
browser = webdriver.Chrome('chromedriver.exe')
#원하는 웹사이트 열기
browser.get('https://www.melon.com/chart/index.htm')

```

```python
#웹페이지 정보 받기 소스를 알려줘!
html = browser.page_source
#정보를 쉽게 찾을 수 있게 bs사용
soup = bs(html, 'html.parser')
```

```python
#원하는 정보를 찾기
#soup.select
#soup.select('태그명')
#soup.select('.class속성값')
#soup.select('#id속성값')
#soup.select('태그명.class속성값')
#soup.select('상위태그 > 하위태그')

#1.노래 정보 뭉치 찾기 -->100개
#2. 노래 1개 의 뭉치-->원하는 정보(제목)만 꺼내기

song_100_list = soup.select('tbody > tr')
len(song_100_list)
```

```python
#노래 100개 뭉치 찾음
#이제 노래 1개 뭉치 찾을 거임
song_1_list = song_100_list[0]
song_1_list
#이 위에 코드가 노래 1개에 대한 모든 정보(작가, 제목, 가수, 넘버 등)
#여기서 노래 제목만 뽑고 싶으면...
```

```python
song_1_list_title = song_1_list.select('div.ellipsis.rank01 > span > a')
#len(song_1_list_title)
#song_1_list_title #대괄호가 쳐져 있음(리스트 안에 들어있음)
#song_1_list_title[0] #리스트에서 "하나밖에 없지만 첫번쨰꺼 고르겠다"이므로 첫번쨰 element가 나오는 거
song_1_list_title[0].text #그첫번쨰 element를 text로 바꿔주겠다 글만가져올게!
```

```python
song_1_list_artist = song_1_list.select('div.ellipsis.rank02 > a')
len(song_1_list_artist)
#song_1_list_artist
song_1_list_artist[0].text
```

```python
song_1_list_album = song_1_list.select('div.ellipsis.rank03 > a')
song_1_list_album[0].text
```

```python
list=[]
rank=1

for i in song_100_list:
    title = i.select('div.ellipsis.rank01 > span > a')[0].text
    artist = i.select('div.ellipsis.rank02 > a')[0].text
    album = i.select('div.ellipsis.rank03 > a')[0].text
    list.append(['melon', rank, title, artist, album])
    rank +=1
    
list
```

```python
pretty_list = pd.DataFrame(list, columns=['음원사이트', '순위', '제목','아티스트','앨범'])
pretty_list
#pretty_list.to_excel('melon_music_top100.xlsx', index=False)
```
