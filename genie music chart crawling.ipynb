```python
import selenium
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
browser = webdriver.Chrome('chromedriver.exe')
browser.get('https://www.genie.co.kr/chart/top200')
```

```python
html = browser.page_source
soup=bs(html, 'html.parser') #전체 페이지
```

```python
song_100_list = soup.select('tbody > tr')
len(song_100_list)
```

```python
song_1_list = song_100_list[0]
song_1_list #1등 노래에 관한 모든 정보
```

```python
song_1_list_title = song_1_list.select('td.info > a.title.ellipsis')
song_1_list_title[0].text.strip()
```

```python
song_1_list_artist = song_1_list.select('td.info > a.artist.ellipsis')
song_1_list_artist[0].text.strip()
```

```python
song_1_list_album = song_1_list.select('td.info > a.albumtitle.ellipsis')
song_1_list_album[0].text.strip()
```

```python
list_1=[]
rank=1

for i in song_100_list:
    title=i.select('td.info > a.title.ellipsis')[0].text.strip()
    artist=i.select('td.info > a.artist.ellipsis')[0].text.strip()
    album=i.select('td.info > a.albumtitle.ellipsis')[0].text.strip()
    list_1.append(['Genie', rank, title, artist, album])
    rank+=1
list_1
```

```python
pretty_list_1=pd.DataFrame(list_1, columns=['음원사이트', '순위', '제목' , '아티스트','앨범'])
pretty_list_1
```
